```python
df.groupby("Supervisor District").count()
```

```python
df.groupby("Supervisor District")["Start Time"].count().reset_index() #group by a column and get a count of another
```

```python
df.groupby(["Supervisor District","Year"])["Start Time"].count().reset_index()
```

```python
by_district_and_year = df.groupby(["Supervisor District","Year"])["Start Time"].count().reset_index() # set the groupby as new dataframe
```

```python
by_district_and_year.rename(columns={"Start Time":"Concert Count"},inplace=True) # rename column
```

```python
by_district_and_year.to_csv("concerts-by-district.csv", index=False) #exprot to csv
```
