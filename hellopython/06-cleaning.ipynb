```python
alt.Chart(per_year, title="Free concerts in Los Angeles County by Year").mark_bar(color="#659055",size = 40).encode(
    x="Year:O",
    y="Concerts:Q",
    tooltip="Concerts"
).properties(width = 400)
```

```python
recent_by_district
```

```python
alt.Chart(recent_by_district, title="Number of concerts in supervisor district for 2021").mark_bar(color="#659055",size = 40).encode(
    x= alt.X("Supervisor District:O", sort = 'y'),
    y= alt.Y("Performing Artist:Q", title = "Number of concerts")
).properties(width = 400)
```
