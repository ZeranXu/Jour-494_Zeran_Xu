```python
def cleanup(x):
    y = x.split("$")
    z = y[1]
    
    return z
```

```python
p =  "$115.00"
```

```python
p.split("$")[1]
```

```python
cleanup("$115.00")
```

```python
cleanup(p)
```

```python
df.price_cleanup = df.price.apply(cleanup)
```
