Request URL to get top tracks for authorized content

```python
trackurl = 'https://api.spotify.com/v1/artists/56oDRnqbIiwx4mymNEv7dS/top-tracks?market=US'
```

```python
r = requests.get(trackurl, headers = head).json()
```

Create a dataframe from json response

```python
tracks = pd.json_normalize(r['tracks'])
```

```python
tracks
```

Create list of track ids

```python
tracklist = tracks['id'].tolist()
```

```python
tracklist
```

```python
trackstring = ','.join(tracklist)
```

```python
trackstring
```

add to a new url

```python
infourl = f"https://api.spotify.com/v1/audio-features?ids={trackstring}"
```

```python
infourl
```

request this new url

```python
infor = requests.get(infourl, headers = head).json()
```

```python
infodf = pd.json_normalize(infor['audio_features'])
```

```python
infodf
```

```python
tracks_simple = tracks[['id', 'name']]
```

Merge trakcs simple with track info database

```python
track_info = pd.merge(tracks_simple, infodf, on = 'id')
```

```python
track_info
```

create a chart of danceablity in Lizzo's top tracks

```python
alt.Chart(track_info, title = "Danceablity in Lizzo's top tracks").mark_bar().encode(
    x = 'danceability:Q',
    y = alt.Y('name:N', sort= '-x')
)
```

```python
alt.Chart(track_info, title="Danceability and energy in Lizzo's top tracks").mark_circle().encode(
    x = 'danceability',
    y = 'energy',
    color = "name"
)
```
