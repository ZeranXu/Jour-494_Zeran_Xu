```python
import mechanize
import pandas as pd
```

```python
url = "https://safetydata.fra.dot.gov/OfficeofSafety/publicsite/summary.aspx"
```

```python
br = mechanize.Browser()
```

```python
br.set_handle_robots(False) # Ignores robots.txt - bad decision!
```

```python
br.open(url)
```

```python
br.select_form(nr=0) #select the form you want to interact with
```

```python
br.form["ctl00$ContentPlaceHolder1$DropDownYear"]=['2021']
```

```python
 # select the year first, get the name of <select> element and then option value
```

```python
br.form["ctl00$ContentPlaceHolder1$ListBoxStats"] = ["r07"]
```

```python
# select the report option - "Total casulaties by state"
```

```python
br.submit()
```

```python
page = br.response().read()
```

```python
table = pd.read_html(page) # list of tables
```
